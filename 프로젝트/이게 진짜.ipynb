```python
name = input('궁금한 지역(시/도)를 입력해주세요 [ex)서울특별시/인천광역시/대전광역시/광주광역시/대구광역시/울산광역시/부산광역시/경기도/강원도/충청북도/충청남도/전라북도/전라남도/경상북도/경상남도] : ')

import csv

#공연장
a = open('공연장.csv')
a = open('공연장.csv', 'r', encoding='UTF8')

a = csv.reader(a)
a = list(a)
ay = []
av = []

for az in a :
    if name in az[2] :
        ay.append(az[2])

for ax in ay : 
    aw = ax.split(' ')
    au = aw[2]
    if au[-1] == '구' :
        av.append(aw[2])
    else :
        av.append(aw[1])
        
#게임장
b = open('게임장.csv')
b = open('게임장.csv', 'r', encoding='UTF8')

b = csv.reader(b)
b = list(b)
by = []
bv = []

for bz in b :
    if name in bz[2] :
        by.append(bz[2])

for bx in by : 
    bw = bx.split(' ')
    bu = bw[2]
    if bu[-1] == '구' :
        bv.append(bw[2])
    else :
        bv.append(bw[1])
        
#노래방
c = open('노래방.csv')
c = open('노래방.csv', 'r', encoding='UTF8')

c = csv.reader(c)
c = list(c)
cy = []
cv = []

for cz in c :
    if name in cz[2] :
        cy.append(cz[2])

for cx in cy : 
    cw = cx.split(' ')
    cu = cw[2]
    if cu[-1] == '구' :
        cv.append(cw[2])
    else :
        cv.append(cw[1])

#당구장
d = open('당구장.csv')
d = open('당구장.csv', 'r', encoding='UTF8')

d = csv.reader(d)
d = list(d)
dy = []
dv = []

for dz in d :
    if name in dz[2] :
        dy.append(dz[2])

for dx in dy : 
    dw = dx.split(' ')
    du = dw[2]
    if du[-1] == '구' :
        dv.append(dw[2])
    else :
        dv.append(dw[1])
        
#영화관
e = open('영화관.csv')
e = open('영화관.csv', 'r', encoding='UTF8')

e = csv.reader(e)
e = list(e)
ey = []
ev = []

for ez in e :
    if name in ez[2] :
        ey.append(ez[2])

for ex in ey : 
    ew = ex.split(' ')
    eu = ew[2]
    if eu[-1] == '구' :
        ev.append(ew[2])
    else :
        ev.append(ew[1])
        
#PC방
f = open('PC방.csv')
f = open('PC방.csv', 'r', encoding='UTF8')

f = csv.reader(f)
f = list(f)
fy = []
fv = []

for fz in f :
    if name in fz[2] :
        fy.append(fz[2])

for fx in fy : 
    fw = fx.split(' ')
    fu = fw[2]
    if fu[-1] == '구' :
        fv.append(fw[2])
    else :
        fv.append(fw[1])
        
count = {}
        
for town in av :
    if town in count :
        count[town] +=1
    else :
        count[town] = 1

for town in bv :
    if town in count :
        count[town] +=1
    else :
        count[town] = 1

for town in cv :
    if town in count :
        count[town] +=1
    else :
        count[town] = 1

for town in dv :
    if town in count :
        count[town] +=1
    else :
        count[town] = 1

for town in ev :
    if town in count :
        count[town] +=1
    else :
        count[town] = 1
        
for town in fv :
    if town in count :
        count[town] +=1
    else :
        count[town] = 1

result = sorted(count.items(), key = lambda t : t[1], reverse=True)
result = list(result)
resultt = [list(result[0])[0],list(result[1])[0],list(result[2])[0]]
print(name,'의 핫플(시/군/구) TOP3 :',resultt)

#공연장
at = []
ar = []

for az in a :
    if resultt[0] in az[2] or resultt[1] in az[2] or resultt[2] in az[2] :
        at.append(az[2])

for ax in at : 
    aw = ax.split(' ')
    au = aw[2]
    if au[-1] == '구' :
        ar.append(aw[3])
    else :
        ar.append(aw[2])
        
#게임장
bt = []
br = []

for bz in b :
    if resultt[0] in bz[2] or resultt[1] in bz[2] or resultt[2] in bz[2] :
        bt.append(bz[2])

for bx in bt : 
    bw = bx.split(' ')
    bu = bw[2]
    if bu[-1] == '구' :
        br.append(bw[3])
    else :
        br.append(bw[2])
        
#노래방
ct = []
cr = []

for cz in c :
    if resultt[0] in cz[2] or resultt[1] in cz[2] or resultt[2] in cz[2] :
        ct.append(cz[2])

for cx in ct : 
    cw = cx.split(' ')
    cu = cw[2]
    if cu[-1] == '구' :
        cr.append(cw[3])
    else :
        cr.append(cw[2])
        
#당구장
dt = []
dr = []

for dz in d :
    if resultt[0] in dz[2] or resultt[1] in dz[2] or resultt[2] in dz[2] :
        dt.append(dz[2])

for dx in dt : 
    dw = dx.split(' ')
    du = dw[2]
    if du[-1] == '구' :
        dr.append(dw[3])
    else :
        dr.append(dw[2])
        
#영화관
et = []
er = []

for ez in e :
    if resultt[0] in ez[2] or resultt[1] in ez[2] or resultt[2] in ez[2] :
        et.append(ez[2])

for ex in et : 
    ew = ex.split(' ')
    eu = ew[2]
    if eu[-1] == '구' :
        er.append(ew[3])
    else :
        er.append(ew[2])
        
#PC방
ft = []
fr = []

for fz in f :
    if resultt[0] in fz[2] or resultt[1] in fz[2] or resultt[2] in fz[2] :
        ft.append(fz[2])

for fx in ft : 
    fw = fx.split(' ')
    fu = fw[2]
    if fu[-1] == '구' :
        fr.append(fw[3])
    else :
        fr.append(fw[2])
        
countt = {}
        
for town in ar :
    if town in countt :
        countt[town] +=1
    else :
        countt[town] = 1

for town in br :
    if town in countt :
        countt[town] +=1
    else :
        countt[town] = 1

for town in cr :
    if town in countt :
        countt[town] +=1
    else :
        countt[town] = 1

for town in dr :
    if town in countt :
        countt[town] +=1
    else :
        countt[town] = 1

for town in er :
    if town in countt :
        countt[town] +=1
    else :
        countt[town] = 1
        
for town in fr :
    if town in countt :
        countt[town] +=1
    else :
        countt[town] = 1

total = sorted(countt.items(), key = lambda t : t[1], reverse=True)
print(name,'의 핫플(읍/면/동) 순위 :',total)
```
